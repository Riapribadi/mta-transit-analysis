##### Using API endpoint -- but the result only shows 2014?

```python
import json
import numpy as np
import pandas as pd
import requests
```

```python
response_API = requests.get('https://data.cityofnewyork.us/resource/ertz-hr4r.json')
print(response_API.status_code)
```

```python
data = response_API.text
```

```python
traffic_count_json = json.loads(data)
traffic_count_json
```

```python
type(traffic_count_json)
```

```python
with open("data/traffic_count.json", "w") as f:
    json.dump(traffic_count_json, f, indent=4) #use json.dump() to output a file 
```

##### Use Pandas to read JSON data

```python
with open("data/traffic_count.json", "r") as f:
    traffic_count = json.load(f) #use json.load() to read the json file


traffic_count
```

```python
df = pd.DataFrame(traffic_count)
df
```
